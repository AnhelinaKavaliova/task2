**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
high_income_men = data[(data['salary'] == '>50K') & (data['sex'] == 'Male')]
married_men_mask = high_income_men['marital-status'].str.startswith('Married')

married_men_frac = high_income_men[married_men_mask].shape[0]/high_income_men.shape[0]
single_men_frac = high_income_men[~married_men_mask].shape[0]/high_income_men.shape[0]

if married_men_frac > single_men_frac:
  print("More men with a salary >50k among married people")
else: print("More men with a salary >50k among single people")
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
def get_age_group(age):
    if 16 <= age < 35:
        return 'young'
    elif 35 <= age < 70:
        return 'adult'
    elif 70 <= age <= 100:
        return 'retiree'
    else:
        return 0

data['AgeGroup'] = data['age'].apply(get_age_group)
data.head()
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
high_income = data[data['salary'] == '>50K']

high_income_agegroups = high_income.groupby('AgeGroup').size()
total_agegroups = data.groupby('AgeGroup').size()
high_income_agegroups_frac = high_income_agegroups/total_agegroups

print("Number of earners >50K in each age group:\n",high_income_agegroups )
print("Age group where people most often earn more than 50K:", high_income_agegroups_frac.idxmax())
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
grouped_occupation = data.groupby('occupation')
occupation_size = grouped_occupation.size()

def filter_func(groups):
  valid_groups = []

  for name_group, group in groups:
    if group['age'].mean() <= 40 and (group['hours-per-week'] > 5).all():
      valid_groups.append(name_group)

  return valid_groups

print("Number of people in each group: \n", occupation_size)
print("Valid groups: ", filter_func(grouped_occupation))
```
